```python
class Node:
  def __init__(self,data):
    self.data = data
    self.next = None
class SingleLinkedList:
  def __init__(self):
    self.head = None
  # this is the function to add new_node at the first of the linked list
  def addnode(self,data):
    new_node = Node(data)
    new_node.next = self.head
    self.head = new_node
  def create_from_arr(self,arr):
    for i in arr:
      if self.head is None:
        new_node = Node(i)
        self.head = new_node
      else:
        current = self.head
        while current.next is not None:
          current = current.next
        new_node = Node(i)
        current.next = new_node
        new_node.next = None
p1 = SingleLinkedList()
p1.create_from_arr([7,6,9])
```

```python
class Treenode:
  def __init__(self,data):
    self.data = data
    self.left = None
    self.right = None

class BinarySearchtree:
  def __init__(self):
    self.root = None

  def insert_key(self,root,key):
    if root is None:
      return Treenode(key)
    if key<root.data:
      root.left = self.insert_key(root.left,key)
    else:
      root.right = self.insert_key(root.right,key)
    return root
  def insert_keys(self,key):
    self.root = self.insert_key(self.root,key)
  """ writing functions to search for the elements in the binary searcg tree"""

  def search_an_element(self,key):
    return self._search_key(self.root,key)
  def _search_key(self,root,key):
    if root is None:
      return "The element is not present in the tree"
    if root.data==key:
      return "The element is present in the key"
    if key<root.data:
      return self._search_key(root.left,key)
    else:
      return self._search_key(root.right,key)
  """yes i have achieved this"""
  # function to find the  maximum element in the tree
  def max_element(self):
    current = self.root
    while current.right is not None:
      current = current.right
    return current.data

  # function to findout the minimum element in the tree

  def min_element(self):
    current = self.root
    while current.left is not None:
      current = current.left
    return current.data
  """"
  Implementation of
  Traversals in Binary search tree

  preorder traversals - root,left,right
  posorder traversals - left,right,root
  inorder traversals - left,root,right
  """
  def inorder_traversal(self):
    result =[]
    self._inorder_traversal(self.root,result)
    return result

  def _inorder_traversal(self,root,result):
    if root is not None:
      self._inorder_traversal(root.left,result)
      result.append(root.data)
      self._inorder_traversal(root.right,result)

  def preorder_traversal(self):
    result = []
    self._preorder_traversal(self.root,result)
    return result

  def _preorder_traversal(self,root,result):
    if root is not None:
      result.append(root.data)
      self._preorder_traversal(root.left,result)
      self._preorder_traversal(root.right,result)

  def postorder_traversal(self):
    result = []
    self._postorder_traversal(self.root,result)
    return result

  def _postorder_traversal(self,root,result):
    if root is not None:
      self._postorder_traversal(root.left,result)
      self._postorder_traversal(root.right,result)
      result.append(root.data)
  # function to find the height of the binary tree
  def height_of_the_tree(self, root):
      if root is None:
        return 0
      else:
        left_height = self.height_of_the_tree(root.left)
        right_height = self.height_of_the_tree(root.right)
        return max(left_height, right_height) + 1
```
